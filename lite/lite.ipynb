### 袛芯屑邪褕薪械械 蟹邪写邪薪懈械 Lite

小芯蟹写邪泄褌械 褋懈褋褌械屑褍 泻芯屑锌褜褞褌械褉薪芯谐芯 蟹褉械薪懈褟, 泻芯褌芯褉邪褟 斜褍写械褌 芯锌褉械写械谢褟褌褜 褌懈锌 谐械芯屑械褌褉懈褔械褋泻芯泄 褎懈谐褍褉褘. 袠褋锌芯谢褜蟹褍褟 锌芯写谐芯褌芯胁谢械薪薪褍褞 斜邪蟹褍 懈 褕邪斜谢芯薪 薪芯褍褌斜褍泻邪 锌褉芯胁械写懈褌械 褋械褉懈褞 褝泻褋锌械褉懈屑械薪褌芯胁 锌芯 锌械褉械斜芯褉褍 谐懈锌械褉锌邪褉邪屑械褌褉芯胁 薪械泄褉芯薪薪芯泄 褋械褌懈, 褉邪褋锌芯蟹薪邪褞褖械泄 褌褉懈 泻邪褌械谐芯褉懈懈 懈蟹芯斜褉邪卸械薪懈泄 (褌褉械褍谐芯谢褜薪懈泻, 泻褉褍谐, 泻胁邪写褉邪褌).

1. 袩芯屑械薪褟泄褌械 泻芯谢懈褔械褋褌胁芯 薪械泄褉芯薪芯胁 胁 褋械褌懈, 懈褋锌芯谢褜蟹褍褟 褋谢械写褍褞褖懈械 蟹薪邪褔械薪懈褟:

- 芯写懈薪 褋谢芯泄 10 薪械泄褉芯薪芯胁
- 芯写懈薪 褋谢芯泄 100 薪械泄褉芯薪芯胁
- 芯写懈薪 褋谢芯泄 5000 薪械泄褉芯薪芯胁.

2. 袩芯屑械薪褟泄褌械 邪泻褌懈胁邪褑懈芯薪薪褍褞 褎褍薪泻褑懈褞 胁 褋泻褉褘褌褘褏 褋谢芯褟褏 褋 `relu` 薪邪 `linear`.
3. 袩芯屑械薪褟泄褌械 褉邪蟹屑械褉褘 batch_size:
- 10
- 100
- 1000

4. 袙褘胁械写懈褌械 薪邪 褝泻褉邪薪 锌芯谢褍褔懈胁褕懈械褋褟 褌芯褔薪芯褋褌懈.

袙褋械谐芯 写芯谢卸薪芯 锌芯谢褍褔懈褌褜褋褟 18 泻芯屑斜懈薪邪褑懈泄 褍泻邪蟹邪薪薪褘褏 锌邪褉邪屑械褌褉芯胁.

小芯蟹写邪泄褌械 褋褉邪胁薪懈褌械谢褜薪褍褞 褌邪斜谢懈褑褍 锌芯 褉械蟹褍谢褜褌邪褌邪屑 锌褉芯胁械写械薪薪褘褏 褌械褋褌芯胁.

```python
from tensorflow import keras
from keras.layers import Dense, Flatten
from keras.models import Sequential
from keras.optimizers import Adam
from keras import utils
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
# 袩褍褌褜 泻 写懈褉械泻褌芯褉懈懈 褋 斜邪蟹芯泄
base_dir = '/content/hw_light'
# 小芯蟹写邪薪懈械 锌褍褋褌芯谐芯 褋锌懈褋泻邪 写谢褟 蟹邪谐褉褍蟹泻懈 懈蟹芯斜褉邪卸械薪懈泄 芯斜褍褔邪褞褖械泄 胁褘斜芯褉泻懈
x_train = []
# 小芯蟹写邪薪懈械 褋锌懈褋泻邪 写谢褟 屑械褌芯泻 泻谢邪褋褋芯胁
y_train = []
# 袟邪写邪薪懈械 胁褘褋芯褌褘 懈 褕懈褉懈薪褘 蟹邪谐褉褍卸邪械屑褘褏 懈蟹芯斜褉邪卸械薪懈泄
img_height = 20
img_width = 20
# 袩械褉械斜芯褉 锌邪锌芯泻 胁 写懈褉械泻褌芯褉懈懈 斜邪蟹褘
for patch in os.listdir(base_dir):
    # 袩械褉械斜芯褉 褎邪泄谢芯胁 胁 锌邪锌泻邪褏
    for img in os.listdir(base_dir + '/' + patch):
        # 袛芯斜邪胁谢械薪懈械 胁 褋锌懈褋芯泻 懈蟹芯斜褉邪卸械薪懈泄 褌械泻褍褖械泄 泻邪褉褌懈薪泻懈
        x_train.append(image.img_to_array(image.load_img(base_dir + '/' + patch + '/' + img,
                                                    target_size=(img_height, img_width),
                                                    color_mode='grayscale')))
        # 袛芯斜邪胁谢械薪懈械 胁 屑邪褋褋懈胁 屑械褌芯泻, 褋芯芯褌胁械褌褋褌胁褍褞褖懈褏 泻谢邪褋褋邪屑
        if patch == '0':
            y_train.append(0)
        elif patch == '3':
            y_train.append(1)
        else:
            y_train.append(2)

# 袩褉械芯斜褉邪蟹芯胁邪薪懈械 胁 numpy-屑邪褋褋懈胁 蟹邪谐褉褍卸械薪薪褘褏 懈蟹芯斜褉邪卸械薪懈泄 懈 屑械褌芯泻 泻谢邪褋褋芯胁
x_train = np.array(x_train)
y_train = np.array(y_train)
# 袙褘胁芯写 褉邪蟹屑械褉薪芯褋褌械泄
print('袪邪蟹屑械褉 屑邪褋褋懈胁邪 x_train', x_train.shape)
print('袪邪蟹屑械褉 屑邪褋褋懈胁邪 y_train', y_train.shape)
```

```python
model = Sequential() # 小芯蟹写邪薪懈械 屑芯写械谢懈

# 袛芯斜邪胁谢械薪懈械 褋谢芯械胁 褋 10 薪械泄褉芯薪邪屑懈, 褎褍薪泻褑懈褟 邪泻褌懈胁邪褑懈懈 relu
model.add(Flatten(input_shape=(20, 20, 1)))
model.add(Dense(10, input_dim = 400, activation='relu'))
model.add(Dense(3))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 袟邪写邪薪懈械 batch_size=100
model.fit(x_train, y_train, epochs=10, batch_size=10)

# 袠蟹屑械褉械薪懈械 褌芯褔薪芯褋褌懈
loss, accuracy = model.evaluate(x_train, y_train)

# 小锌懈褋芯泻 泻芯屑斜懈薪邪褑懈泄 锌邪褉邪屑械褌褉芯胁
results = []
results.append([10, "relu", 10, accuracy])
model.save_weights('lite.weights.h5')
```
